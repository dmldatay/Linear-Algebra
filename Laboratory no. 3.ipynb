```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$$v_A = c_1 (-x + 2y) \\ v_B = c_2 (4x   -2y) \\  $$
$$v_A = c_1 \cdot\begin{bmatrix} -1\\2 \\\end{bmatrix}, 
v_B = c_2 \cdot \begin{bmatrix} 4\\-2 \\\end{bmatrix}$$

```python
vectA = np.array([-1,2])
vectB = np.array([4,-2])

c1 = np.arange(-10,10,1)
c2 = np.arange(-10,10,1)
plt.scatter(c1*vectA[0],c1*vectA[1])
plt.scatter(c2*vectB[0],c2*vectB[1])

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')


plt.grid()
plt.show()
```

<h2><b>Task 2</b></h2>

$$A = c_1 (-2x + y) \\ B = c_2 (x - 3y) \\  $$
$$S = \begin{Bmatrix} c_1 \cdot\begin{bmatrix} -2\\1 \\\end{bmatrix}, 
c_2 \cdot \begin{bmatrix} 1\\-3 \\\end{bmatrix}\end{Bmatrix} $$

```python
vectA = np.array([-2,1])
vectB = np.array([1,-3])
Q = np.arange(-10,10,0.5)
C1, C2 = np.meshgrid(Q,Q)
vectQ = vectA + vectB
spanQx = C1*vectA[0] + C2*vectB[0]
spanQy = C1*vectA[1] + C2*vectB[1]
plt.scatter(spanQx,spanQy, s=5, alpha=0.75)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```
